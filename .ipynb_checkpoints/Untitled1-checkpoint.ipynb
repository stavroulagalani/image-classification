```python
#LOAD DATA
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = tf.keras.utils.image_dataset_from_directory( r"C:\image classification CNN\data"  )
```

```python
data_iterator = data.as_numpy_iterator()
```

```python
#SCALE DATA
data = data.map(lambda x, y: (x/255, y))
```

```python
data.as_numpy_iterator().next()[0].max()
```

```python
len(data)
```

```python
train_size = int(len(data)*.7) #5
val_size = int(len(data)*.2)+1 #2
test_size = int(len(data)*.1)+1 #1
```

```python
train = data.take(train_size)
val = data.skip(train_size).take(val_size)
test = data.skip(train_size+val_size).take(test_size)
```

```python
#DEEP MODEL
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten
```

```python
model = Sequential()
```

```python
#convolution = takes input and extracts importand features, activation func(relu) = implements non-linearity for more comlplex patterns, maxpooling = reduces size,keeping most importand info
model.add(Conv2D(16, (3,3), 1, activation='relu', input_shape=(256,256,3)))
model.add(MaxPooling2D())

model.add(Conv2D(32, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())

model.add(Conv2D(16, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())

model.add(Flatten()) #from 3d converts it to 1d

model.add(Dense(256, activation='relu')) #connects the layers and makes predictions
model.add(Dense(1, activation='sigmoid')) #sigmoid because of linearity gives 0 or 1,if multi-layer 'softmax' outputs a probability distribution
```

```python
model.compile('adam', loss=tf.losses.BinaryCrossentropy(), metrics=['accuracy'])
```

```python
model.summary()
```

```python
#TRAIN THE MODEL
logdir = 'logs'
```

```python
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir = logdir)
```

```python
history = model.fit(train, epochs=25, validation_data = val, callbacks = [tensorboard_callback])
```

```python
history.history
```

```python
#plot the performance
fig = plt.figure()
plt.plot(history.history['loss'], color='teal', label='loss')
plt.plot(history.history['val_loss'], color='orange', label='val_loss')
fig.suptitle('Loss', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

```python
fig = plt.figure()
plt.plot(history.history['accuracy'], color='teal', label='accuracy')
plt.plot(history.history['val_accuracy'], color='orange', label='val_accuracy')
fig.suptitle('Accuracy', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

```python
#EVALUATE PERFORMANCE
from tensorflow.keras.metrics import Precision, Recall, BinaryAccuracy
```

```python
pre = Precision()
re = Recall()
acc = BinaryAccuracy()
```

```python
for batch in test.as_numpy_iterator(): 
    X, y = batch
    yhat = model.predict(X)
    pre.update_state(y, yhat)
    re.update_state(y, yhat)
    acc.update_state(y, yhat)
```

```python
print(f'Precision:{pre.result().numpy()}, Recall:{re.result().numpy()}, Accuracy:{acc.result().numpy()}')
```
